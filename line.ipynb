```python
from datetime import datetime
import pandas as pd

DATE_FORMAT = '%a, %m/%d/%Y'

print(datetime.now().strftime(DATE_FORMAT))

with open('[LINE] Chat in Epic Chat Chain.txt', 'r') as f:
    fulltext = f.read()
    days = fulltext.split('\n\n')

# day 0 is header
for day in days[1:4]:
    lines = day.split('\n')
    date = datetime.strptime('%a, %m/%d/%Y', lines[0])
    print(date)
```
